```python
import pandas as pd
import numpy as np 
import datetime as dt
```

```python
####Reading the order dataset
orders = pd.read_csv("Input/orders.csv")
orders["order_ready_at"] = pd.to_datetime(orders['order_ready_at'], format='%Y-%m-%d %H:%M:%S.%f')
orders["order_acknowledged_at"] = pd.to_datetime(orders['order_acknowledged_at'], format='%Y-%m-%d %H:%M:%S.%f')
orders["prep_time_minutes"] = (orders["order_ready_at"] - orders["order_acknowledged_at"]).dt.total_seconds() / 60.0
orders
```

```python
####Reading the restaurants dataset
rest = pd.read_csv("Input/restaurants.csv")
rest
```

```python
####Understanding the datasets
print("Total number of unique restaurants in the order dataset =",len(orders["restaurant_id"].unique()))
print("\nTotal number of unique restaurants in the restaurant dataset =",len(rest["restaurant_id"].unique()))
print("\nUnique countries in the restaurant dataset =",rest["country"].unique())
print("\nTotal number of unique cities in the restaurant dataset =",len(rest["city"].unique()),"\nCities: ",rest["city"].unique())
print("\nTotal number of unique type of food in the restaurant dataset =",len(rest["type_of_food"].unique()))
```
